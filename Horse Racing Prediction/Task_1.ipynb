## 1.Numerical_df => Pick all those distances which are less than equal to 1800. 

```python
df_4 = df_3[df_3['distance'] <= 1800]
```

## 2. Numerical_df => Pick all those horse career earnings which are greater than equal to 90000. 

```python
df_5 = df_4[df_4['horse_career_earnings'] <= 90000]
```

## 3. Numerical_df => Pick all those total which are greater than equal to 450002
 

```python
df_6 = df_5[df_5['total'] >= 450002]
```

## 4. Numerical_df => Pick all positions which are (top 5). 

```python
df_7 = df_6[df_6['position'] <= 5]
```

## 5. Remove those columns which holds 20% or greater null values.


```python
df_8 = df_7.isnull().mean()
df_8
```

```python
df_9 = df_7.drop(['dam_by', 'stud', 'body_weight', 'incident1', 'allowance', 'ex_name', 'bit1'], axis = 1)
```
