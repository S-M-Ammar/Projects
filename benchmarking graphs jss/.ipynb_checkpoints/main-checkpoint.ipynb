```python
import os
from os import walk
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
ROOT_DIR = os.path.abspath(os.curdir)

files = []
for (dirpath, dirnames, filenames) in walk(ROOT_DIR+"/records"):
    files.extend(filenames)
   
```

```python
len(files)
```

```python
for file_ in files:
    df = pd.read_csv(ROOT_DIR+"/records/"+file_,skiprows=8)
    plt.figure(figsize=(20.0,10.0))
    plt.ylim(1150,1450)
    plt.plot( df['Generation #'],df['Run-1'], linewidth=2)
    plt.plot( df['Generation #'],df['Run-2'], linewidth=2)
    plt.plot( df['Generation #'],df['Run-3'], linewidth=2)
    plt.plot( df['Generation #'],df['Run-4'], linewidth=2)
    plt.plot( df['Generation #'],df['Run-5'], linewidth=2)
    plt.legend(['Run-1','Run-2','Run-3','Run-4','Run-5'])
    plt.xlabel("Generations")
    plt.ylabel("MakeSpan")
    plt.savefig(ROOT_DIR+"/graph_pics/"+file_.replace('csv',"")+"png")
```
