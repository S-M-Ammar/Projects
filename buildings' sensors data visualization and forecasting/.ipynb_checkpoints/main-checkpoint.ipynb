```python
# For forecasting, we need to check either data is stationary or not.
# For this, we use dicky fuller test.
```

```python
from statsmodels.tsa.stattools import adfuller
def addfuller_test(data):
    result = adfuller(data)
    labels = ['ADF Test Statistic','p-value','# Lags Used','# Observations Used']
    for value,label in zip(result,labels):
        print(label+' : '+str(value))
    if(result[1]<= 0.05):
        print("Data is stationary")
    else:
        print("Data is non stationary")
```

```python
# For No2, we will set lookback = 10
# For Co2, we will set lookback = 10
# For PM2.5, we will set lookback = 8
```

```python
import plaidml.keras
import os
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Flatten
import math
from sklearn.metrics import mean_squared_error
plaidml.keras.install_backend()
```

```python
# function to create supervised time-series dataset for LSTM 
def create_dataset(dataset, time_step=1):
    dataX, dataY = [], []
    for i in range(len(dataset)-time_step-1):
        a = dataset[i:(i+time_step), 0]   ###i=0, 0,1,2,3-----99   100 
        dataX.append(a)
        dataY.append(dataset[i + time_step, 0])
    return np.array(dataX), np.array(dataY)
```

```python
def forecasting(df,lookback):
    df1 = df
    scaler=MinMaxScaler(feature_range=(0,1))
    df1=scaler.fit_transform(np.array(df1).reshape(-1,1))
    
    ##splitting dataset into train and test split
    training_size=int(len(df1)*0.60)
    test_size=len(df1)-training_size
    train_data,test_data=df1[0:training_size,:],df1[training_size:len(df1),:1]
    
    # reshape into X=t,t+1,t+2,t+3 and Y=t+4
    time_step = lookback
    X_train, y_train = create_dataset(train_data, time_step)
    X_test, ytest = create_dataset(test_data, time_step)
    
    # reshape input to be [samples, time steps, features] which is required for LSTM
    X_train =X_train.reshape(X_train.shape[0],X_train.shape[1] , 1)
    X_test = X_test.reshape(X_test.shape[0],X_test.shape[1] , 1)
    
    # Preparing model
    model=Sequential()
    model.add(LSTM(50,return_sequences=True,input_shape=(time_step,1)))
    model.add(LSTM(50,return_sequences=True))
    model.add(LSTM(50))
    model.add(Dense(1))
    model.compile(loss='mean_squared_error',optimizer='adam')
    model.fit(X_train,y_train,validation_data=(X_test,ytest),epochs=200,batch_size=len(X_train)-1,verbose=1)
    
    
    # check mean squared error
    test_predict=model.predict(X_test)
    test_predict=scaler.inverse_transform(test_predict)
    mse = mean_squared_error(ytest,test_predict)
    return model,mse,test_data
    

```

```python
from numpy import array
def forecast_next_values(x_input,temp_input,lookback,model):
    # demonstrate prediction for next 10 days
    

    lst_output=[]
    n_steps=lookback
    i=0
    while(i<10):
    
        if(len(temp_input)>n_steps):
            #print(temp_input)
            x_input=np.array(temp_input[1:])
    #         print("{} day input {}".format(i,x_input))
            x_input=x_input.reshape(1,-1)
            x_input = x_input.reshape((1, n_steps, 1))
            #print(x_input)
            yhat = model.predict(x_input, verbose=0)
    #         print("{} day output {}".format(i,yhat))
            temp_input.extend(yhat[0].tolist())
            temp_input=temp_input[1:]
            #print(temp_input)
            lst_output.extend(yhat.tolist())
            i=i+1
        else:
            x_input = x_input.reshape((1, n_steps,1))
            yhat = model.predict(x_input, verbose=0)
    #         print(yhat[0])
            temp_input.extend(yhat[0].tolist())
    #         print(len(temp_input))
            lst_output.extend(yhat.tolist())
            i=i+1
    
    return lst_output
```

#### No2 Model

```python
no2 = df_sensor_A_avg[['NO2 (ppm)']]
no2 = no2.reset_index()['NO2 (ppm)']
no2_model,mse,test_data = forecasting(no2,10)
```

```python
# Now lets see results for no2 model
```

```python
print("Mean sqaured value ",mse)
```

```python
x_input=test_data[0:10].reshape(1,-1) # Here 10 is lookback
temp_input=list(x_input)
temp_input=temp_input[0].tolist()
temp_input
forecast = forecast_next_values(x_input,temp_input,10,no2_model)
forecast = np.abs(forecast)
print("Next 10 Days value for No2\n")
print(forecast)
```

```python
# As we can see, no2 model is really effective as it has 0.21 mse. 
```
