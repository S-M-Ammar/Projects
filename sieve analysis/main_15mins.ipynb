```python
# configurations
# For probe-1 => spez Oberfl. [cm²/g] = 123.07 , MK calc [mm] = 0.20 , AFS = 69 , MK [mm] = 0.272 , GG[%] = 80.50
# For probe-2 => spez Oberfl. [cm²/g] = 121.63 , MK calc [mm] = 0.20 , AFS = 68 , MK [mm] = 0.28 , GG[%] = 79.85
# For probe-3 => spez Oberfl. [cm²/g] = 124.48 , MK calc [mm] = 0.19 , AFS = 70 , MK [mm] = 0.27 , GG[%] = 80.04
```

```python
def get_configs(a,b,c,d,e):
    spez = []
    Mk_calc = []
    AFS = []
    MK = []
    GG = []
    for x in range(len(df_1)):
        spez.append(a)
        Mk_calc.append(b)
        AFS.append(c)
        MK.append(d)
        GG.append(e)

    return spez,Mk_calc,AFS,MK,GG   
```

```python
spez,Mk_calc,AFS,MK,GG   = get_configs(123.07,0.20,69,0.272,80.50) # for probe-1
df_1['spez Oberfl. [cm²/g]'] = spez
df_1['MK calc [mm]'] = Mk_calc
df_1['AFS'] = AFS
df_1['MK [mm]'] = MK
df_1['GG [%]'] = GG

spez,Mk_calc,AFS,MK,GG   = get_configs(121.63,0.20,68,0.28,79.85) # for probe-2
df_2['spez Oberfl. [cm²/g]'] = spez
df_2['MK calc [mm]'] = Mk_calc
df_2['AFS'] = AFS
df_2['MK [mm]'] = MK
df_2['GG [%]'] = GG

spez,Mk_calc,AFS,MK,GG   = get_configs(124.48,0.19,70,0.27,80.04) # for probe-3
df_3['spez Oberfl. [cm²/g]'] = spez
df_3['MK calc [mm]'] = Mk_calc
df_3['AFS'] = AFS
df_3['MK [mm]'] = MK
df_3['GG [%]'] = GG
```

```python
df_1.fillna(value=0,inplace=True)
df_2.fillna(value=0,inplace=True)
df_3.fillna(value=0,inplace=True)
```

```python
# getting final dataframe
final_df = pd.concat([df_1,df_2,df_3],ignore_index=True)
final_df
```

```python
# check correlation
correlation_data = final_df.corr()
cols_to_be_selected = []
all_cols = correlation_data.columns
index = 0;

for x in correlation_data['Massenanteil [%]'].tolist():
    if(x >= -0.25 and x <=0.25):
        #neglecting the feature / column
        pass
    else:
        cols_to_be_selected.append(all_cols[index])
    index += 1

cols_to_be_selected
```

```python
features = final_df[['do',
 'du',
 'Tara',
 'Massenanteil [g]',
 'g*M1',
 'g*M2',
 'g*M3']]

target = final_df[['Massenanteil [%]']]

```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_validate
reg = LinearRegression()
model = cross_validate(reg,features,target,cv=10,scoring='r2',return_estimator=True)
model
```

```python
final_model = model['estimator'][0]
final_model.coef_[0]
```

```python
final_model.intercept_
```
