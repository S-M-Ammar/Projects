```python
df_1 = df_1.iloc[:,0:total_columns-2]
df_2 = df_2.iloc[:,0:total_columns-2]
df_3 = df_3.iloc[:,0:total_columns-2]
df_1
```

```python
# configurations
# For probe-1 => spez Oberfl. [cm²/g] = 124.91 , MK calc [mm] = 0.19 , AFS = 70 , MK [mm] = 0.2696 , GG[%] = 80.35
# For probe-1 => spez Oberfl. [cm²/g] = 126.20 , MK calc [mm] = 0.19 , AFS = 71 , MK [mm] = 0.27 , GG[%] = 79.94
# For probe-1 => spez Oberfl. [cm²/g] = 124.98 , MK calc [mm] = 0.19 , AFS = 70 , MK [mm] = 0.27 , GG[%] = 80.24
```

```python
def get_configs(a,b,c,d,e):
    spez = []
    Mk_calc = []
    AFS = []
    MK = []
    GG = []
    for x in range(len(df_1)):
        spez.append(a)
        Mk_calc.append(b)
        AFS.append(c)
        MK.append(d)
        GG.append(e)

    return spez,Mk_calc,AFS,MK,GG   
```

```python
spez,Mk_calc,AFS,MK,GG   = get_configs(124.91,0.19,70,0.2696,80.35) # for probe-1
df_1['spez Oberfl. [cm²/g]'] = spez
df_1['MK calc [mm]'] = Mk_calc
df_1['AFS'] = AFS
df_1['MK [mm]'] = MK
df_1['GG [%]'] = GG

spez,Mk_calc,AFS,MK,GG   = get_configs(126.20,0.19,71,0.27,80.24) # for probe-2
df_2['spez Oberfl. [cm²/g]'] = spez
df_2['MK calc [mm]'] = Mk_calc
df_2['AFS'] = AFS
df_2['MK [mm]'] = MK
df_2['GG [%]'] = GG

spez,Mk_calc,AFS,MK,GG   = get_configs(124.98,0.19,70,0.27,80.35) # for probe-3
df_3['spez Oberfl. [cm²/g]'] = spez
df_3['MK calc [mm]'] = Mk_calc
df_3['AFS'] = AFS
df_3['MK [mm]'] = MK
df_3['GG [%]'] = GG
```

```python
df_1.fillna(value=0,inplace=True)
df_2.fillna(value=0,inplace=True)
df_3.fillna(value=0,inplace=True)
```

```python
# getting final dataframe
final_df = pd.concat([df_1,df_2,df_3],ignore_index=True)
final_df
```

```python
# check correlation
correlation_data = final_df.corr()
cols_to_be_selected = []
all_cols = correlation_data.columns
index = 0;

for x in correlation_data['Massenanteil [%]'].tolist():
    if(x >= -0.25 and x <=0.25):
        #neglecting the feature / column
        pass
    else:
        cols_to_be_selected.append(all_cols[index])
    index += 1

cols_to_be_selected
```

```python
features = final_df[['do',
 'du',
 'Tara',
 'Massenanteil [g]',
 'g*M1',
 'g*M2',
 'g*M3']]

target = final_df[['Massenanteil [%]']]

```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_validate
reg = LinearRegression()
model = cross_validate(reg,features,target,cv=10,scoring='r2',return_estimator=True)
model
```

```python
final_model = model['estimator'][0]
final_model.coef_[0][6]
```

```python
final_model.intercept_
```
