# Importing required libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib 
from matplotlib import pyplot as plt
%matplotlib inline
sns.set(rc={'figure.figsize':(20.7,10.27)})
```

# Reading data file for all covid-19 policies 

```python
#reading dataset
df_policies = pd.read_excel('acaps_covid19_government_measures_dataset_0.xlsx',sheet_name="Dataset")
df_policies
```

# Getting data for united states only

```python
#getting all records for US (United States)
df_policies = df_policies[df_policies['COUNTRY']=="United States"]
df_policies
```

# Cleaning & filtering the data

```python
# getting useful columns
df_policies = df_policies[['COUNTRY','LOG_TYPE','CATEGORY','MEASURE','DATE_IMPLEMENTED']]
df_policies
```

```python
#removing null data
df_policies.dropna(axis=0,inplace=True)
df_policies
```

```python
# sorting data in ascending order with respect to "date implemented" column
df_policies.sort_values(by=['DATE_IMPLEMENTED'],ascending=True,inplace=True)

# renaming "DATE IMPLEMENTED" column
df_policies.rename(columns={'DATE_IMPLEMENTED':'DATE'},inplace=True)
```

# Types of logs for policies

```python
for x in df_policies['LOG_TYPE'].unique():
    print("\n",x,"\n")
```

# Types of categories for policies

```python
for x in df_policies['CATEGORY'].unique():
    print("\n",x,"\n")
```

# Types of measure for policies

```python
for x in df_policies['MEASURE'].unique():
    print("\n",x,"\n")
```

# Lets see which category of policies implemented most

```python
categorical_policies_count = df_policies['CATEGORY'].value_counts()
sns.barplot(x=categorical_policies_count.index,y=categorical_policies_count.values)
```

# Lets see which measure of policies implemented most

```python
measure_policies_count = df_policies['MEASURE'].value_counts()
sns.barplot(x=measure_policies_count.values,y=measure_policies_count.index)
```

# Reading covid-19 dataset of United States

```python
df_covid = pd.read_csv('us_covid19_daily.csv')
df_covid
```

# Selecting useful features 

```python
df_covid = df_covid[['lastModified','totalTestResults','positive','negative','hospitalizedCumulative','onVentilatorCumulative','recovered','death','deathIncrease','hospitalizedIncrease','negativeIncrease','positiveIncrease']]
df_covid
```

```python
# converting "last modified" column into date format.
df_covid['lastModified'] = df_covid['lastModified'].astype('datetime64[ns]')

# arranging dataset with rspect to date in ascending order
df_covid.sort_values(by=['lastModified'],ascending=True,inplace=True)

# renaming "last modified" column
df_covid.rename(columns={'lastModified':'DATE'},inplace=True)

# filling null values / records
df_covid.fillna(value=0,inplace=True)
df_covid
```

```python
copy_covid = df_covid
copy_policies = df_policies
```

# Now First we need to see the effect of policies w.r.t categories on covid dataset in terms of deaths, recovered etc.

# For this we need to perform some grouping between two datasets

```python
# getting first record for every category
policies_category_records = pd.DataFrame() 
for x in list(categorical_policies_count.index):
    record = copy_policies[copy_policies['CATEGORY']==x].head(1)
    policies_category_records = policies_category_records.append(record,ignore_index=True)
    
```

```python
#showing records
policies_category_records
```

# Now we can easily see that 

## Public Health Measure implemented on 7/January/2020

## Governance & Socio-Ecnomice Measures implmented on 21/January/2020

## Movement Restrictions implemented on 31/January/2020

## Social distancing implemented on 10/March/2020

```python
# getting last record for every category
policies_category_records_last = pd.DataFrame() 
copy_policies_out = copy_policies[copy_policies['LOG_TYPE']=="Phase-out measure"]
for x in list(categorical_policies_count.index):
    record = copy_policies_out[copy_policies_out['CATEGORY']==x].tail(1)
    policies_category_records_last = policies_category_records_last.append(record,ignore_index=True)
    
```

# Lets see the effects of  Public Health Measures & Governance & Socio-Ecnomice Measures as the covid records starts from 22/January/2020

```python
# As the movement restrictions policy implemented on 31st Januaury. So we need data before 31st January to check tht stats for our above two policies.

stats_1 = copy_covid[copy_covid['DATE']<='2020-01-31']
stats_1['DATE'] = stats_1['DATE'].astype('string')
stats_1
```

# We can see that there are no important records. Now we will add our third policy which is Movement Restrictions

```python
# We have collected down the records for first three policies. 
# 1. Public health measures
# 2. Governance & Socio-Ecnomice Measures
# 3. Movement Restriction

# The social distancing applied on 10th march 2020. So we need data before that to see the impact of first 3 policies.
stats_2 = copy_covid[(copy_covid['DATE']>'2020-01-31') & (copy_covid['DATE']<='2020-03-10') ]
stats_2['DATE'] = stats_2['DATE'].astype('string')
stats_2
```

```python
hospitalized = stats_2.tail(1)['hospitalizedCumulative']
hospitalized = hospitalized.values
hospitalized = int(round(float(hospitalized),0))
print("Patients in hospital = ",hospitalized)

on_ventilator = stats_2.tail(1)['onVentilatorCumulative']
on_ventilator = on_ventilator.values
on_ventilator = int(round(float(on_ventilator),0))
print("Patients on ventilator = ",on_ventilator)

total_positive = stats_2['positive'].sum() 
total_negative = stats_2['negative'].sum() 
total_result = stats_2['totalTestResults'].sum() 
```

```python

# Data to plot
labels = ['total_positive','total_negative']
sizes = [total_positive,total_negative]
colors = ['gold','lightcoral']
explode = (0.1, 0)  # explode 1st slice

# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=True, startangle=140)

plt.axis('equal')
plt.title("Ratio of positive and negative covid-19 patients For first 3 policies",size=20)
plt.show()
```

```python
plt.title('Ratio of hospitalized and On Ventilator covid-19 patients For first 3 policies',size=20)
sns.barplot(x=['hospitalizedCumulative','onVentilatorCumulative'],y=[hospitalized,on_ventilator])
```

## Lets see some line graphs to get the ups and downs of stats.

```python
total_length = len(stats_2)
plt.title('Death Chart For first 3 policies',size=20)
plt.xlabel('Days')
plt.ylabel('Deaths')

sns.lineplot(x=[i for i in range(1,total_length+1)],y=stats_2['death'])
```

```python
mutliGraph = stats_2[['deathIncrease','hospitalizedIncrease','negativeIncrease','positiveIncrease']]
mutliGraph['days'] = [i for i in range(1,total_length+1)]
sns.lineplot('days', 'value', hue='variable', data=pd.melt(mutliGraph, 'days'))
```

## So it can be clearly seen that in initial days, the policies have not much effect but later on, the effect is highlighted as there is a good number of "negative increases" going up. Despite of this, also number of deaths got high too in later days. We can assume that might be the patients that are effected from covid-19 before policies implemented. (Just an assumption)
