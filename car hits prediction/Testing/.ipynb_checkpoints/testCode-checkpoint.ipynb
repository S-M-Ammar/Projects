```python
import pandas as pd
import numpy as np
import pickle
from numpy import absolute
```

```python
df = pd.read_csv('STOCKVO_GTC_20201115.csv',sep=";",encoding='latin1')
```

```python
features = df[['CO2','KMS']]
```

```python
model = pickle.load(open('final_model.pkl', 'rb'))
```

```python
preds = model.predict(features)
```

```python
preds = absolute(preds)
```

```python
hits = []
for x in preds:
    try:
        hits.append(round(x[0],15))
    except:
        hits.append(x[0])
```
