```python
import pandas as pd
import pickle
import numpy as np
import requests
import json
import matplotlib.pyplot as plt
pd.set_option('display.max_rows', 200)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', -1)
```

```python
df = pd.read_csv('final_dataset.csv',index_col=0)
df
```

```python
def Convert_possession(x):
    num = int(x[0:len(x)-1])
    return (num/100)
    
df['Away_Ball_Poss'] = df['Away_Ball_Poss'].apply(lambda x:Convert_possession(x))
df['Home_Ball_Poss'] = df['Home_Ball_Poss'].apply(lambda x:Convert_possession(x))
df.fillna(value=0,inplace=True)

df = df.loc[:,'Away_Team_Name':]
df
```

```python
# attacking_strength of Home & Away

attacking_strength_home = []
attacking_strength_away = []
defensive_strength_home = []
defensive_strength_away = []


for i in range(0,len(df)):
    
    #attacking strength
    Home_Shots_Inside_Box = df.iloc[i]['Home_Shots_InsideBox']
    Away_Shots_Inside_Box = df.iloc[i]['Away_Shots_InsideBox']
    Total_Shots_Inside_Box = Home_Shots_Inside_Box + Away_Shots_Inside_Box
    
    Home_Shots_Off_Goal = df.iloc[i]['Home_shots_OFFGoal']
    Away_Shots_Off_Goal = df.iloc[i]['Away_shots_OFFGoal']
    Total_Shots_Off_Goal= Home_Shots_Off_Goal + Away_Shots_Off_Goal
    
    Home_Shots_On_Goal= df.iloc[i]['Home_shots_ONGoal']
    Away_Shots_On_Goal= df.iloc[i]['Away_shots_ONGoal']
    Total_Shots_On_Goal= Home_Shots_On_Goal + Away_Shots_On_Goal
    
    Home_Shots_Outside_Box = df.iloc[i]['Home_shots_OutSideBox']
    Away_Shots_Outside_Box = df.iloc[i]['Away_shots_OutSideBox']
    Total_Shots_Outside_Box = Home_Shots_Outside_Box + Away_Shots_Outside_Box
    
    Home_corner_kicks = df.iloc[i]['Home_Corner_Kicks']
    Away_corner_kicks = df.iloc[i]['Away_Corner_Kicks']
    Total_corner_kicks = df.iloc[i]['Home_Corner_Kicks'] + Away_corner_kicks
    
    attack_strength_home = ((Home_Shots_Inside_Box/Total_Shots_Inside_Box)*100)+((Home_Shots_Off_Goal/Total_Shots_Off_Goal)*100)+((Home_Shots_On_Goal/Total_Shots_On_Goal)*100)+((Home_Shots_Outside_Box/Total_Shots_Outside_Box)*100)+((Home_corner_kicks/Total_corner_kicks)*100)
    attack_strength_home = attack_strength_home / 500
    
    
    attack_strength_away = ((Away_Shots_Inside_Box/Total_Shots_Inside_Box)*100)+((Away_Shots_Off_Goal/Total_Shots_Off_Goal)*100)+((Away_Shots_On_Goal/Total_Shots_On_Goal)*100)+((Away_Shots_Outside_Box/Total_Shots_Outside_Box)*100)+((Away_corner_kicks/Total_corner_kicks)*100)
    attack_strength_away = attack_strength_away / 500
  
    
    attacking_strength_home.append(round(attack_strength_home,2))
    attacking_strength_away.append(round(attack_strength_away,2))
    
    #defensive strength
    Home_blocked_shots = df.iloc[i]['Home_Blocked_Shots']
    Away_blocked_shots = df.iloc[i]['Away_Blocked_Shots']
    Total_shots_blocked = Home_blocked_shots + Away_blocked_shots
    
    Home_keeper_saves = df.iloc[i]['Home_Keeper_Saves']
    Total_for_Home = df.iloc[i]['Away_Total_Shots'] - df.iloc[i]['Away_shots_OFFGoal']
    
    Away_keeper_saves = df.iloc[i]['Away_Keeper_Saves']
    Total_for_Away = df.iloc[i]['Home_Total_Shots'] - df.iloc[i]['Home_shots_OFFGoal']
    
    Home_Interceptions = df.iloc[i]['Away_Total_Passes'] - df.iloc[i]['Away_Passes_Accurate']
    Total_pass_for_home = df.iloc[i]['Away_Total_Passes']
    
    Away_Interceptions = df.iloc[i]['Home_Total_Passes'] - df.iloc[i]['Home_Passes_Accurate']
    Total_pass_for_away = df.iloc[i]['Home_Total_Passes']
    
    defense_strength_home = ((Home_blocked_shots/Total_shots_blocked)*100)+((Home_keeper_saves/Total_for_Home)*100)+((Home_Interceptions/Total_pass_for_home)*100)
    defense_strength_home = defense_strength_home/300
    
    defense_strength_away = ((Away_blocked_shots/Total_shots_blocked)*100)+((Away_keeper_saves/Total_for_Away)*100)+((Away_Interceptions/Total_pass_for_away)*100)
    defense_strength_away = defense_strength_away/300
    
    defensive_strength_home.append(round(defense_strength_home,2))
    defensive_strength_away.append(round(defense_strength_away,2))
    
```

```python
df['Attacking_Strength_Home'] = attacking_strength_home 
df['Attacking_Strength_Away'] = attacking_strength_away
df['Defensive_Strength_Home'] = defensive_strength_home
df['Defensive_Strength_Away'] = defensive_strength_away
df = df.loc[:,'Away_Ball_Poss':]
df.fillna(value=0,inplace=True)
df
```

# Predicting Home Goals

```python

# For Home Goals, data training & testing data preparation
from sklearn.model_selection import train_test_split
target = df['Home_Goals']
features = df.drop(['Home_Goals','Away_Goals'],axis=1)

train_features, test_features, train_labels, test_labels = train_test_split(features, target, test_size = 0.25, random_state = 42)

print('Training Features Shape:', train_features.shape)
print('Training Labels Shape:', train_labels.shape)
print('Testing Features Shape:', test_features.shape)
print('Testing Labels Shape:', test_labels.shape)
```

```python
from sklearn import linear_model
from sklearn.metrics import r2_score,mean_squared_error,mean_squared_log_error

regr = linear_model.Lasso(alpha=0.1)
regr = regr.fit(train_features, train_labels)
preds = regr.predict(test_features)
preds = np.abs(np.round(preds,0))
print("R2 Score ",r2_score(test_labels, preds))
print("MSE ",mean_squared_error(test_labels, preds))
print("MSLE ",mean_squared_log_error(test_labels, preds))

# save the model to disk
filename = 'home_goals_model.sav'
pickle.dump(regr, open(filename, 'wb'))
```

# Predicting Away Goals

```python
# For Away Goals, data training & testing data preparation
from sklearn.model_selection import train_test_split
target = df['Away_Goals']
features = df.drop(['Home_Goals','Away_Goals'],axis=1)

train_features, test_features, train_labels, test_labels = train_test_split(features, target, test_size = 0.25, random_state = 42)

print('Training Features Shape:', train_features.shape)
print('Training Labels Shape:', train_labels.shape)
print('Testing Features Shape:', test_features.shape)
print('Testing Labels Shape:', test_labels.shape)
```

```python
regr = linear_model.Lasso(alpha=0.1)
regr = regr.fit(train_features, train_labels)
preds = regr.predict(test_features)
preds = np.abs(np.round(preds,0))
print("R2 Score ",r2_score(test_labels, preds))
print("MSE ",mean_squared_error(test_labels, preds))
print("MSLE ",mean_squared_log_error(test_labels, preds))

# save the model to disk
filename = 'away_goals_model.sav'
pickle.dump(regr, open(filename, 'wb'))
```
