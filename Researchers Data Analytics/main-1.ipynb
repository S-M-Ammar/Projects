```python
from pmdarima.arima import auto_arima
from pmdarima.arima import ADFTest
import numpy as np
import pandas as pd
pd.set_option('display.max_rows', 800)
pd.set_option('display.max_columns', 500)
```

```python
df = pd.read_excel('./Predictive Forecasting Items for Fiverr .xlsx',sheet_name="Wins & Losses ",skiprows=4)
df
```

```python
df.fillna(value=0,inplace=True)
df
```

```python
unique_ids = df['SponsorPIAffiliateID'].unique()
unique_researchers = df['SponsorPIFullName'].unique()
unique_researchers
```

```python
adf_test = ADFTest(alpha=0.5)
```

# Forecasting for NSF Category

```python
NSF_Wins_df = pd.DataFrame(columns=['SponsorPIAffiliateID','SponsorPIFullName','NSF Year-1 Forecast-Wins','NSF Year-2 Forecast-Wins','NSF Year-3 Forecast-Wins','NSF Year-4 Forecast-Wins','NSF Year-5 Forecast-Wins'])
NSF_Loss_df = pd.DataFrame(columns=['SponsorPIAffiliateID','SponsorPIFullName','NSF Year-1 Forecast-Loss','NSF Year-2 Forecast-Loss','NSF Year-3 Forecast-Loss','NSF Year-4 Forecast-Loss','NSF Year-5 Forecast-Loss'])
NSF_Withdrawn_df = pd.DataFrame(columns=['SponsorPIAffiliateID','SponsorPIFullName','NSF Year-1 Forecast-Withdrawn','NSF Year-2 Forecast-Withdrawn','NSF Year-3 Forecast-Withdrawn','NSF Year-4 Forecast-Withdrawn','NSF Year-5 Forecast-Withdrawn'])
counter = 0
NSF_df = df[["Year","SponsorPIAffiliateID","SponsorPIFullName","NSFWins","NSF$Won","NSFLosses","NSF$Losses","NSFWithdrawn","NSF$Withdrawn"]]
for x in unique_researchers:
    temp_df = NSF_df[NSF_df['SponsorPIFullName']==x]
    temp_df.sort_values(by="Year",ascending=True,inplace=True)
#     test_result = adf_test.should_diff(temp_df[['NSFWins']].to_numpy())[1]
    print(temp_df['NSFWins'].tolist())
    arima_nsf_wins = auto_arima(temp_df['NSFWins'],start_p=0,d=0,start_q=0,max_p=5,max_d=5,max_q=5,start_P=0,D=1,start_Q=0,
                                max_P=5,max_D=5,max_Q=5,m=12,seasonal=False,error_Action='warn',trace=True,supress_warnings=True,stepwise=True,random_state=20,n_fits=50)
    
    arima_nsf_loss = auto_arima(temp_df['NSFLosses'],start_p=0,d=0,start_q=0,max_p=5,max_d=5,max_q=5,start_P=0,D=1,start_Q=0,
                                max_P=5,max_D=5,max_Q=5,m=12,seasonal=False,error_Action='warn',trace=True,supress_warnings=True,stepwise=True,random_state=20,n_fits=50)

    arima_nsf_withdrawn = auto_arima(temp_df['NSFWithdrawn'],start_p=0,d=0,start_q=0,max_p=5,max_d=5,max_q=5,start_P=0,D=1,start_Q=0,
                                max_P=5,max_D=5,max_Q=5,m=12,seasonal=False,error_Action='warn',trace=True,supress_warnings=True,stepwise=True,random_state=20,n_fits=50)
    
    
#     print(arima_nsf_wins.summary())
#     print(arima_nsf_loss.summary())
#     print(arima_nsf_withdrawn.summary)
    
    wins_pred = np.absolute(arima_nsf_wins.predict(n_periods=5))
    loss_pred = np.absolute(arima_nsf_loss.predict(n_periods=5))
    withdrawn_pred = np.absolute(arima_nsf_withdrawn.predict(n_periods=5))
    
    
    NSF_Wins_df.loc[len(NSF_Wins_df.index)] = [unique_ids[counter],x,round(wins_pred[0],0),round(wins_pred[1],0),round(wins_pred[2],0),round(wins_pred[3],0),round(wins_pred[4],0)] 
    NSF_Loss_df.loc[len(NSF_Loss_df.index)] = [unique_ids[counter],x,round(loss_pred[0],0),round(loss_pred[1],0),round(loss_pred[2],0),round(loss_pred[3],0),round(loss_pred[4],0)] 
    NSF_Withdrawn_df.loc[len(NSF_Withdrawn_df.index)] = [unique_ids[counter],x,round(withdrawn_pred[0],0),round(withdrawn_pred[1],0),round(withdrawn_pred[2],0),round(withdrawn_pred[3],0),round(withdrawn_pred[4],0)] 
    counter=counter+1
    
    
    

```

```python
NSF = NSF_Wins_df.merge(NSF_Loss_df,on='SponsorPIFullName')
NSF = NSF.merge(NSF_Withdrawn_df,on='SponsorPIFullName')
NSF.drop(labels=['SponsorPIAffiliateID_y','SponsorPIAffiliateID'],axis=1,inplace=True)
NSF.rename(columns={'SponsorPIAffiliateID_x': 'SponsorPIAffiliateID'}, inplace=True)
NSF
```

# Optimal Funding Amount & Win Ratio Per Category

```python
df_historic = pd.read_excel('./Predictive Forecasting Items for Fiverr .xlsx',sheet_name="Historic",skiprows=6)
counter = 0

df_Win_Ratio = pd.DataFrame(columns=['ID','Name','NSF Win Chance','NIH Win Chance','Other Win Chance','NSF Optimal Fund','NIH Optimal Fund','Other Optimal Fund'])

for x in unique_researchers:
    temp_df = df[df['SponsorPIFullName']==x]
    temp_df.sort_values(by="Year",ascending=True,inplace=True)
    temp_df = temp_df[['NSFWins','NIHWins','OtherWins','NSF$Won','NIH$Won','Other$Won']]
    
    nsf_wins = temp_df[['NSFWins']].sum()[0]
    nih_wins = temp_df[['NIHWins']].sum()[0]
    other_wins = temp_df[['OtherWins']].sum()[0]
    
    nfs_max_fund = temp_df[['NSF$Won']].max()[0]
    nih_max_fund = temp_df[['NIH$Won']].max()[0]
    other_max_fund = temp_df[['Other$Won']].max()[0]
    
    data = df_historic[df_historic['ID']==unique_ids[counter]]

    if(len(data)>=1):
        
        total_attempts = data.iloc[0][4]
        nsf_perc = (nsf_wins/total_attempts)*100
        nih_perc = (nih_wins/total_attempts)*100
        other_perc = (other_wins/total_attempts)*100
        df_Win_Ratio.loc[len(df_Win_Ratio)] = [unique_ids[counter],x,nsf_perc,nih_perc,other_perc,nfs_max_fund,nih_max_fund,other_max_fund]
    
    counter=counter+1
    
 
```
