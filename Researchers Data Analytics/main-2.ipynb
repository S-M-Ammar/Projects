```python
from pmdarima.arima import auto_arima
from pmdarima.arima import ADFTest
import numpy as np
import pandas as pd
pd.set_option('display.max_rows', 800)
pd.set_option('display.max_columns', 500)
```

```python
df = pd.read_excel('./Predictive Forecasting Items for Fiverr .xlsx',sheet_name="Publications",skiprows=0)
df
```

```python
df.fillna(value=0,inplace=True)
df
```

```python
unique_researchers = df['PI Name'].unique()
unique_researchers
```

```python
pubs_df = pd.DataFrame(columns=['Name','2021','2022','2023','2024','2025'])
for x in unique_researchers:
    temp_df = df[df['PI Name']==x]
    temp_df = temp_df.iloc[:,4:]
    data = []
    for k in range(0,12):
        data.append(temp_df.iloc[0][k])
    
    arima_pubs = auto_arima(data,start_p=0,d=0,start_q=0,max_p=10,max_d=10,max_q=10,start_P=0,D=1,start_Q=0,
                                max_P=5,max_D=5,max_Q=5,m=12,seasonal=False,error_Action='warn',trace=True,supress_warnings=True,stepwise=True,random_state=20,n_fits=50)
    
    pubs_pred = np.absolute(arima_pubs.predict(n_periods=5))
    final_preds = []
    for y in pubs_pred:
        final_preds.append(round(y,0))
    
    
    pubs_df.loc[len(pubs_df.index)] = [x,final_preds[0],final_preds[1],final_preds[2],final_preds[3],final_preds[4]]
    
  
```

```python
pubs_df.to_csv('Publications Forecasting.csv')
```
