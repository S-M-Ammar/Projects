```python
import pandas as pd
import numpy as np
pd.set_option('display.max_rows', 800)
pd.set_option('display.max_columns', 500)

import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from nba_api.stats.static import teams

nba_teams = teams.get_teams()
team_ = [team for team in nba_teams if team['full_name'] == 'Los Angeles Lakers'][0]
team_id = team_['id']
```

```python
from nba_api.stats.endpoints import leaguegamefinder

# Query for games where the Celtics were playing
gamefinder = leaguegamefinder.LeagueGameFinder(team_id_nullable=team_id)
# The first DataFrame of those returned is what we want.
games = gamefinder.get_data_frames()[0]
games.head()
```

```python
# convert season id into year id
def convert_season_id(x):
    return x[-4:]

games['SEASON_ID'] = games['SEASON_ID'].apply(lambda x:convert_season_id(x))
```

```python
season_records_df = games.groupby(games.SEASON_ID.str[-4:])[['GAME_ID']].count().loc[:]

season_records_df
```

```python
# getting total number of seasons,
total_number_of_seasons = len(season_records_df.index)
print("Total number of season",total_number_of_seasons)

# getting last 3 seasons for fetching latest stats / performance
season_records_df = season_records_df.tail(5)
season_records_df
```

```python
# getting all games record for latest season
seasons_record = season_records_df.index.tolist()

games = games[(games['SEASON_ID']==seasons_record[0]) | (games['SEASON_ID']==seasons_record[1]) | (games['SEASON_ID']==seasons_record[2])]
games
```
