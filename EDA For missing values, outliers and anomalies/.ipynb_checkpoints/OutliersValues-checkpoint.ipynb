# Importing Libraries


```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib
%matplotlib inline
```

# Reading Outliers Data

```python
df = pd.read_csv('data/OutlierData (Updated).csv')
df
```

```python
df.info()
```

# Using box plot &  IQR (interquartile range) For Outliers


### Checking order_price

```python
sns.boxplot(df['order_price'])
```

### We can clearly see that all order_price values which are greater than 35000 or less than zero, must be fall in category of outliers.

```python
# Removing outliers 
df = df[df['order_price']<=35000]
df = df[df['order_price']>0]
df
```

### Checking delivery charges

```python
sns.boxplot(df['delivery_charges'])
```

### We can clearly see that all delivery chargers that are greater then 115 and less than 40 will be considered as outliers

```python
# Removing outliers 
df = df[df['delivery_charges']<=115]
df = df[df['delivery_charges']>=40]
df
```

### Checking coupon discount

```python
sns.boxplot(df['coupon_discount'])
```

### There are no outliers in coupon feature


### Checking order_total

```python
sns.boxplot(df['order_total'])
```

### We can see that order_total values that are greater than 31000 will be considered as outliers

```python
# Removing outliers 
df = df[df['order_total']<=31000]
df = df[df['order_total']>=2]
df
```

### Checking distance_to_nearest_warehouse

```python
sns.boxplot(df['distance_to_nearest_warehouse'])
```

### Distance that is greater than 2.35 is considered as outlier

```python
# Removing outliers 
df = df[df['distance_to_nearest_warehouse']<=2.35]
df
```

### Checking customer latitude & longitude

```python
sns.boxplot(df['customer_lat'])
```

```python
sns.boxplot(df['customer_long'])
```

### No outlier detected

### Saving output 

```python
df.to_csv('Outlier-output.csv')
```
